```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'area':['A city','B city','C city'],'pets_missing':[10,15,9],'std':[3,6,2],'dogs':[5,7,3],'cats':[5,8,6]})
```

```python
df
```

# std data

```python
df.pets_missing.std()
```

# single bar chart

```python
plt.bar(df.area,df.pets_missing)
plt.ylabel('pets_missing')
plt.show()
```

# stacked bar chart

```python
plt.bar(df.area,df.dogs,label='Dogs')
plt.bar(df.area,df.cats,bottom=df.dogs,label='Cats')
plt.legend()
plt.show()
```

```python
plt.barh(df.area,df.pets_missing)
plt.xlabel('pets_missing')
plt.show()
```

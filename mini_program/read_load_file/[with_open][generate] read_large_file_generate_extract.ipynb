# simple generate function

```python
def num_squence(n):
    i=0
    while i<n:
        yield i
        i += 1
```

```python
list(num_squence(3))
```

# open file with generate function

```python
#define read_large_file()
def read_large_file(file_object):
    """A generator function to read a large file lazily."""
    
    #generate loop (keep if data not end)
    while True:
        #read line function
        data = file_object.readline()
        #break (break if data is end)
        if not data:
            break
        #set yield (return)
        yield data
```

```python
with open('WDI_csv/WDIData.csv') as file:
    gen_file = read_large_file(file)
    print(next(gen_file))
    print(next(gen_file))
    print(next(gen_file))
```

# dict pivot, with generate, using for loop view lines
> previous practice, use next to print some of lines

> how about using for loop view all lines, like sql scanning all values

```python
count_dict = {}
```

```python
with open('WDI_csv/WDIData.csv') as file:
    for line in read_large_file(file):
        #extract process
        row = line.split(',')
        first_col = row[0]
        
        #dict process
        if first_col in count_dict.keys():
            count_dict[first_col] += 1
        else:
            count_dict[first_col] = 1
```
